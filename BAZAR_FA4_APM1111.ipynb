<strong> 1) Find the (a) first, (b) second, (c) third, and (d) fourth moments for each of the sets of data (normal, skewed-right, skewed-left, uniform). </strong>

```python
import pandas as pd

def calculateRawMoments(data):
    moments = {}
    n = len(data)  

    moments['First Moment (Mean)'] = data.sum() / n

    moments['Second Moment'] = (data ** 2).sum() / n

    moments['Third Moment'] = (data ** 3).sum() / n

    moments['Fourth Moment'] = (data ** 4).sum() / n

    return moments

normalData = data['Normal']
skewedRightData = data['Skewed-right']
skewedLeftData = data['Skewed-left']
uniformData = data['Uniform']

normalRawMoments = calculateRawMoments(normalData)
skewedRightRawMoments = calculateRawMoments(skewedRightData)
skewedLeftRawMoments = calculateRawMoments(skewedLeftData)
uniformRawMoments = calculateRawMoments(uniformData)

rawMomentsTable = pd.DataFrame({
    'Normal': normalRawMoments.values(),
    'Skewed-right': skewedRightRawMoments.values(),
    'Skewed-left': skewedLeftRawMoments.values(),
    'Uniform': uniformRawMoments.values()
}, index=['First Moment (Mean)', 'Second Moment', 'Third Moment', 'Fourth Moment'])


rawMomentsTable
```

<strong> 2) Find the (a) first, (b) second, (c) third, and (d ) fourth moments about the mean for each of the sets of data (normal, skewed-right, skewed-left, uniform). </strong> 

```python
def calculateMomentsAboutMean(data):
    moments = {}
    mean = data.mean() 
    n = len(data) 

    first_moment = round(((data - mean).sum()) / n)
    moments['FirstMoment'] = first_moment

    variance = ((data - mean) ** 2).sum() / n
    moments['SecondMoment'] = variance

    third_moment = ((data - mean) ** 3).sum() / n
    moments['ThirdMoment'] = third_moment

    fourth_moment = ((data - mean) ** 4).sum() / n
    moments['FourthMoment'] = fourth_moment

    return moments

normalData = data['Normal']
skewedRightData = data['Skewed-right']
skewedLeftData = data['Skewed-left']
uniformData = data['Uniform']

normalMoments = calculateMomentsAboutMean(normalData)
skewedRightMoments = calculateMomentsAboutMean(skewedRightData)
skewedLeftMoments = calculateMomentsAboutMean(skewedLeftData)
uniformMoments = calculateMomentsAboutMean(uniformData)

momentsTable = pd.DataFrame({
    'Normal': normalMoments.values(),
    'Skewed-right': skewedRightMoments.values(),
    'Skewed-left': skewedLeftMoments.values(),
    'Uniform': uniformMoments.values()
}, index=['First Moment', 'Second Moment', 'Third Moment', 'Fourth Moment'])

momentsTable
```

<strong> 3) Find the (a) first, (b) second, (c) third, and (d ) fourth moments about the number 75 for the set of female height measurements. </strong>

*Since the question explicitly refers to the set of female height measurements, we only calculate the moments for the normal distribution.*

```python
def calculateMomentsAbout75(data):
    moments = {}
    N = len(data) 
    
    firstmoment = ((data - 75).sum()) / N
    moments['First Moment'] = firstmoment

    secondmoment = (((data - 75) ** 2).sum()) / N
    moments['Second Moment'] = secondmoment

    thirdmoment = (((data - 75) ** 3).sum()) / N
    moments['Third Moment'] = thirdmoment

    fourthmoment = (((data - 75) ** 4).sum()) / N
    moments['Fourth Moment'] = fourthmoment

    return moments

heightData = data['Normal']

heightMoments = calculateMomentsAbout75(heightData)

momentsTable = pd.DataFrame.from_dict(heightMoments, orient='index', columns=['Moments about 75'])

momentsTable
```
